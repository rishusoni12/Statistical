```python
#Line Marenge 
#line graph me dono me hi number chahiye  Function plot
x = [1,2,3,4,5,6]
y = [10,20,30,40,50,60]
plt.plot(x,y, color="r", linestyle="dashed", marker="o", label="y=x^2")  #marker change kar sakte hai linestyle color legend 
plt.xlabel("Student")  
plt.ylabel("Stats Mark")
plt.title("Line plot") #Title de sakte hai 

plt.legend() # label batata hai 
plt.show()
```

```python
x = [1,2,3,4,5,6, 7, 8, 9, 12, 13, 14]
y = [10,20,30,40,50,60, 68,90, 45, 56, 54,68]
plt.scatter(x,y, color="r",linestyle="dashed", marker="*", label="y=x^2")  #marker change kar sakte hai linestyle color legend 
plt.xlabel("Student")  
plt.ylabel("Stats Mark")
plt.title("Scatter plot") #Title de sakte hai 
plt.grid()
plt.legend() # label batata hai 
plt.show()
```

```python
x = [1,2,3,4,5,6, 7, 8, 9, 12, 13, 14]
y = [10,20,30,40,50,60, 68,90, 45, 56, 54,68]
size=[10,100,50,60,800,90,150,80,90, 80, 60, 500]
plt.scatter(x,y, color="r", s=size, linestyle="dashed", marker="*", label="y=x^2")  #marker change kar sakte hai linestyle color legend 
plt.xlabel("Student")  
plt.ylabel("Stats Mark")
plt.title("Scatter plot") #Title de sakte hai 
plt.grid(True, linestyle="dashed", alpha=0.8)
plt.legend() # label batata hai 
plt.show()
```

```python
#histogram
data = np.random.randn(100)
plt.hist(data, bins=25, color="r", alpha=1, edgecolor="black", label="mark") #bin graph size 
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Basic Histogram")
plt.grid()
plt.legend(loc=1)
plt.show()
```

```python
data = np.random.randn(100)
np.random.seed(40) #Fix karta hai random number ko 
plt.hist(data, bins=25, color="r", alpha=1, edgecolor="black", label="mark") #bin graph size 
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Basic Histogram")
plt.grid()
plt.legend(loc=1)
plt.show()
```

```python
import matplotlib.pyplot as plt

sizes = [40, 50, 60, 70]
labels = ["Group1", "Group2", "Group3", "Group4"]
colors = ["blue", "green", "yellow", "orange"]

plt.pie(sizes, labels=labels, autopct="%1.1f%%", startangle=140, colors=colors)
plt.title("Pie chart Example")
plt.show()
```

```python
#Pie Chart
sizes=[40,50,60,70]
labels=["Group1", "Group2", "Group3", "Group4"]
colors=["blue", "green", "yellow", "orange"]
plt.pie(sizes, labels=labels, autopct="%1.1f%%", startangle=140, colors=colors, explode=(0.3,0,0,0))  
plt.title("Pie chart Example")
plt.show()
```

```python
w=0.3 #width
x= ["C1", "C2", "C3", "C4"]
boys=[100,40,50,30]
girls=[200,70,80,50]
bar1=np.arange(len(x)) #kitne bar banage 
plt.bar(bar1,boys,w,label="boys")
bar2=[i+w for i in bar1]
plt.bar(bar2, girls, w, label="girl")
plt.xticks(bar1+w/2,x)
plt.show()
```
